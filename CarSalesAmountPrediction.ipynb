```python
#import libraries
#table manipulations
import pandas as pd
#numerical analysis
import numpy as np
#plots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
car_df = pd.read_csv('Car_Purchasing_Data.csv', encoding='ISO-8859-1')
#encoding of special characters is important
```

```python
#visualising data
car_df
```

```python
sns.pairplot(car_df)
#for data visualisation 
```

```python
#data cleaning
#drop certain features like customer_name,email,country as they are not important and drop output Car Purchase Amount
#inputs
X = car_df.drop(['Customer Name', 'Customer e-mail', 'Country','Car Purchase Amount'], axis = 1)
#axis=1 means drop columns, axis=0 means drop rows
```

```python
X
```

```python
#Outputs
y=car_df['Car Purchase Amount']
```

```python
y
```

```python
X.shape
#no. of rows and no. of columns
```

```python
y.shape
```

```python
#normalising input data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#MinMaxScaler() performs feature scaling transforming values of the dataset into range 0 to 1
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
scaler.data_max_
#stores the maximum value of each feature (column) in dataset X
```

```python
#stores the minimum value of each feature (column) in dataset X
scaler.data_min_
```

```python
#reshape output data as its expecting 2D array but instead got 1D array
y = y.values.reshape(-1,1)
```

```python
y.shape
```

```python
#normalising output data
scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y)
```

```python
y_scaled
```

```python
#divide data into training data and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled,test_size = 0.3)
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler

model = Sequential()
model.add(Dense(25, input_dim=5, activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs=20, batch_size=25,  verbose=1, validation_split=0.2)
```

```python
print(epochs_hist.history.keys())
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])

plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training and Validation Losses')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])
```
